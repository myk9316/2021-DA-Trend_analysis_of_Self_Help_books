```python
import numpy as np
import pandas as pd
import requests as rq
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import requests
from selenium.common.exceptions import NoSuchElementException
import time
import re
```

```python
url='http://www.yes24.com/24/category/bestseller?CategoryNumber=001001046&sumgb=06'
```

```python
driver = webdriver.Chrome(executable_path=r'C:\MIN\1. Python\chromedriver_win32\chromedriver.exe')
driver.get(url)
time.sleep(3)
```

```python
#로그인 먼저하하구 성인인증하기 #안하면 19금 책에서 막힘;;
```

```python
driver.get(url)
book_titles = list()
book_reviews = list() 
for m in list(range(20)):    
    yes = driver.find_element_by_xpath('/html/body/div/div[2]/div[2]/table/tbody/tr[' + str(m*2 + 1) +']/td[3]/p[3]/a[1]').click() #책클릭
    time.sleep(0.5)
    yes = driver.find_element_by_css_selector('#yDetailTopWrap > div.topColRgt > div.gd_infoTop > div > h2')
    title = yes.text
    print('제목: '+ title)
    yes = driver.find_element_by_css_selector('#chkOrderReviewForReview')
    yes.click() #구매리뷰
    try:
        for k in list(range(20)):
            for j in list(range(10)):
                time.sleep(2)
                for i in list(range(5)):
                    yes = driver.find_element_by_css_selector('#infoset_reviewContentList > div:nth-child('+ str(i+3) + ') > div.reviewInfoBot.crop > a > div > span > em')
                    yes.click()
                    time.sleep(0.5)

                for i in list(range(5)):
                    print(i+1)
                    yes = driver.find_element_by_xpath('//*[@id="infoset_reviewContentList"]/div['+ str(i+2) +']/div[3]/div[2]')
                    print(yes.text)
                    print()
                    book_titles.append(title)
                    book_reviews.append(yes.text)    
                print('-------------------------------------페이지 '+ str(k) + str(j+1) + ' 끝---------------------------------------')
                yes = driver.find_element_by_css_selector('#infoset_reviewContentList > div.review_sort.sortBot > div.review_sortLft > div > a:nth-child(' + str(j+4) +')')
                yes.click()
                time.sleep(3)
    except NoSuchElementException:                
        driver.back()
        time.sleep(2)
```
